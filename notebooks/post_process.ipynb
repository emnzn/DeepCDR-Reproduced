```python
import os

import pandas as pd
```

```python
data_dir = os.path.join("..", "data", "cleaned")

train_path = os.path.join(data_dir, "train.csv")
val_path = os.path.join(data_dir, "validation.csv")
test_path = os.path.join(data_dir, "test.csv")
```

```python
train = pd.read_csv(train_path)
val = pd.read_csv(val_path)
test = pd.read_csv(test_path)
```

```python
train.groupby(["cell_line_id", "drug_id"])["cancer_type"].count()
```

```python
train["cell_line_id"].values
```

```python
train_pair = set([f"{c}-{d}" for c, d in zip(train["cell_line_id"].values, train["drug_id"].values)])

train_pair
```

```python
val_pair = set([f"{c}-{d}" for c, d in zip(val["cell_line_id"].values, val["drug_id"].values)])

val_pair
```

```python
sum([int(t in val_pair) for t in train_pair])
```

```python
train_drugs = set(train["drug_id"].values)
val_drugs = set(val["drug_id"].values)
test_drugs = set(test["drug_id"].values)
```

```python
train_d_in_val = sum([int(t in val_drugs) for t in train_drugs])
train_d_in_test = sum([int(t in test_drugs) for t in train_drugs])

print("train drugs in val:", train_d_in_val)
print("train drugs in test:", train_d_in_test)
```

```python
train_cell_lines = set(train["cell_line_id"].values)
val_cell_lines = set(val["cell_line_id"].values)
test_cell_lines = set(test["cell_line_id"].values)
```

```python
train_c_in_val = sum([int(t in val_cell_lines) for t in train_cell_lines])
train_c_in_test = sum([int(t in test_cell_lines) for t in train_cell_lines])

print("train cell lines in val:", train_c_in_val)
print("train cell lines in test:", train_c_in_test)
```
